```python
#CHANGING THE RADIUS TO 30KM IN ORDER TO HAVE THE FULL RANGE OF THE ISLAND!!

def getNearbyVenues(names, latitudes, longitudes, radius=30000):
    
    #MODIFY THE LIMIT UP TO 200 IN ORDER TO GET MORE VENUES!!!
    LIMIT = 200 # limit of number of venues returned by Foursquare API
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

     #Adding Columns   
    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Island', 
                  'Island Latitude', 
                  'Island Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
#making a huge Dataframe with the function getnearbyvenues,which gets the popular venues from Foursquare and also adding the name of the islands,latlong,venues

aegean_venues = getNearbyVenues(names=df['Aegean_Islands'],
                                   latitudes=df['Latitude'],
                                   longitudes=df['Longitude']
                                  )

```

# Data Analysis for potential Tourism Investments based on current Popular Venues. 


```python

df_freq=pd.DataFrame(columns=['Venues','Frequency'])

for row in aegean_grouped['Island']:
    temp = aegean_grouped[aegean_grouped['Island'] == row].T.reset_index()
    temp.columns = ['Venues','Frequency']
    temp=temp.drop(temp.index[0])
    temp=temp.sort_values('Frequency', ascending=False).reset_index(drop=True).head()
    df_freq=df_freq.append(temp, ignore_index=True)
```

# Clustering the Islands.

```python
print(' INITIALIZE THE K MEANS,WITH 10 CENTROIDS NOT ONLY BECAUSE OF THE DISTANCE BETWEEN THE ISLAND,BUT ALSO BECAUSE WE WANT TO HAVE BETTER RESULTS WITH MORE PERCISION')
```

```python
#!!!
# INITIALIZE THE K MEANS,WITH 10 CENTROIDS NOT ONLY BECAUSE OF THE DISTANCE BETWEEN THE ISLAND
#BUT ALSO BECAUSE WE WANT TO HAVE BETTER RESULTS with more percision
kclusters = 10

aegean_grouped_clustering = aegean_grouped.drop('Island', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(aegean_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

```python
aegean_merged = df

# add clustering labels
aegean_merged['Cluster Labels'] = kmeans.labels_

#the last version of aegean Islands with the addition of Cluster Labels!
#notice that the cluster labels can go from 0 up to 9 !!
aegean_merged 
```

# Clustering the Islands into 10 groups,tagged with Labels.

```python
# create map
map_clusters = folium.Map(location=[latitude, longitude], zoom_start=8)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i+x+(i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, jas, cluster in zip(aegean_merged['Latitude'], aegean_merged['Longitude'], aegean_merged['Aegean_Islands'], aegean_merged['Cluster Labels']):
    label = folium.Popup(str(jas) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
