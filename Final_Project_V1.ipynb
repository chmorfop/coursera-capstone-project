```python
from bs4 import BeautifulSoup
import requests

url="https://en.wikipedia.org/wiki/List_of_Aegean_Islands"
r = requests.get(url)
soup = BeautifulSoup(r.text, 'html.parser')  



datasoup = soup.find('table')
#print(datasoup)
```

```python

Names=[]


for row in datasoup.find_all('tr'):
     for cell in row.find_all('td'):
            if (cell.find('a', class_='new')):
                break
            name=cell.find('b').text 
            if(name=="Christiana" or name=="Pachia"):
                break
            Names.append(name)
            break
                
                
print(Names)


       
```

```python
ClearNames=[]

for onoma in Names:
    if(onoma.find("(")):
        onoma=onoma.split("(")[0]
        onoma=onoma + ",Greece"
        ClearNames.append(onoma)
        
print(ClearNames) 
```

```python
def getNearbyVenues(names, latitudes, longitudes, radius=15000):
    
    LIMIT = 100 # limit of number of venues returned by Foursquare API
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Island', 
                  'Island Latitude', 
                  'Island Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
aegean_venues = getNearbyVenues(names=df['Aegean_Islands'],
                                   latitudes=df['Latitude'],
                                   longitudes=df['Longitude']
                                  )

```
